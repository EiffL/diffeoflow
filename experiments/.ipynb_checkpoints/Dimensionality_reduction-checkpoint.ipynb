```python
import gzip, zipfile, tarfile
import os, shutil, re, string, urllib, fnmatch
import pickle as pkl

def _download_mnist_realval(dataset):
    """
    Download the MNIST dataset if it is not present.
    :return: The train, test and validation set.
    """
    origin = (
        'http://www.iro.umontreal.ca/~lisa/deep/data/mnist/mnist.pkl.gz'
    )
    print('Downloading data from %s' % origin)
    urllib.request.urlretrieve(origin, dataset)

def _get_datafolder_path():
    full_path = os.path.abspath('.')
    path = full_path +'/data'
    return path

def load_mnist_realval(
        dataset=_get_datafolder_path()+'/mnist_real/mnist.pkl.gz'):
    '''
    Loads the real valued MNIST dataset
    :param dataset: path to dataset file
    :return: None
    '''
    if not os.path.isfile(dataset):
        datasetfolder = os.path.dirname(dataset)
        if not os.path.exists(datasetfolder):
            os.makedirs(datasetfolder)
        _download_mnist_realval(dataset)

    f = gzip.open(dataset, 'rb')
    train_set, valid_set, test_set = pkl.load(f, encoding='latin1')
    f.close()
    x_train, targets_train = train_set[0], train_set[1]
    x_valid, targets_valid = valid_set[0], valid_set[1]
    x_test, targets_test = test_set[0], test_set[1]
    return x_train, targets_train, x_valid, targets_valid, x_test, targets_test
```

```python
x_train, targets_train, x_valid, targets_valid, x_test, targets_test = load_mnist_realval()
```

```python
# Define the flow
def func(y, t, stage):
    with tf.variable_scope('network_%d'%stage, reuse=tf.AUTO_REUSE):
        y = tf.layers.dense(y, 512, activation=tf.nn.leaky_relu)
        y = tf.layers.dense(y, 784, activation=tf.nn.leaky_relu)
    return y


def inv_func(y,t,stage):
    return -func(y,t,stage)
```

```python
from functools import partial
```

```python
tf.reset_default_graph()
batch_size = 128
nsteps = 7

lr = tf.placeholder_with_default(0.001,shape=[])
x = tf.placeholder(shape=[None, 784],dtype=tf.float32)
dt = tf.cast(tf.placeholder_with_default(0.01,shape=[]), dtype=tf.float64)
```

```python
# Encoder/Decoder for MNIST image with a flow
inter = []
for i in range(nsteps):
    if i == 0:
        net = odeint_fixed(partial(func, stage=i), x, array([i/(nsteps*1.0), (i+1)/(nsteps*1.0)]).astype('float32'), dt=dt)[-1]
    else:
        net = odeint_fixed(partial(func, stage=i), net,array([i/(nsteps*1.0), (i+1)/(nsteps*1.0)]).astype('float32'), dt=dt)[-1]
    inter.append(net[:, 784//(2**(i+1)):])
    mask = np.zeros((batch_size, 784), dtype='float32')
    mask[:,:784//(2**(i+1))] = 1
    net = net * mask
    print(i, [i/(nsteps*1.0), (i+1)/(nsteps*1.0)])
        
print(2**(i+1),784//(2**(i+1)) )
code = net
print('toto')
for i in range(nsteps):
    net = odeint_fixed(partial(inv_func, stage=nsteps-i-1), net, array([i/(nsteps*1.0), (i+1)/(nsteps*1.0)]).astype('float32'), dt=dt)[-1]
    print(nsteps-i-1)
rec = net
```

```python
loss = tf.reduce_sum([tf.reduce_mean(tf.reduce_sum(p**2,axis=1)) for p in inter]) + tf.reduce_mean(tf.reduce_sum((rec - x)**2,axis=1))
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate=0.001)
```

```python
opt_op = optimizer.minimize(loss)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
inds = choice(len(x_train),batch_size)
c,r,t = sess.run([code,rec,x],feed_dict={x: x_train[inds], lr:0.001})
```
